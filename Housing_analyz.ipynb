```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['AGE'], df['MEDV'])
ax.set_xlabel('AGE')
ax.set_ylabel('MEDV')
plt.show()
```
